```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

#Plotting the data
plt.plot(x, y)

#adding a title, axis labels, and a grid
plt.title("Simple line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()

```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color='yellowgreen')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Student Marks')
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = []
for val in revenue:
    if val == max(revenue):
        explode.append(0.1)
    else:
        explode.append(0)

plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title("Revenue Distribution by Region")
plt.show()
```

```python
data = np.random.randint(1,101,1000)

plt.hist(data , bins=20, color='purple')
plt.title("Histogram of random numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
